```python
import numpy as np
import ot  # pip install POT
from sklearn.metrics import pairwise_distances
from soap import *
```

```python
_, structure, species = read_cif("CIF_files/DB0-m1_o14_smb_repeat.cif")
```

```python
soap_out, soap = S(structure = structure, species = species)
```

```python
soap_out
```

```python
N_env, N_feat = soap_out.shape
D = pairwise_distances(soap_out, metric='sqeuclidean')
```

```python
a = np.ones(N_env) / N_env  # shape = (N_env,)
b = np.ones(N_env) / N_env  # same for the “reference” being itself
```

```python
reg = 1e-1  # entropic regularization strength; you can tune this if desired
P = ot.sinkhorn(a, b, D, reg)  # shape = (N_env, N_env)
```

```python
w = P.sum(axis=1)          # shape = (N_env,)
w /= w.sum()               # normalize so sum(w) = 1
```

```python
d_rematch = w.reshape(1, N_env) @ soap_out  # shape = (1, N_feat)
```

```python
print("d_rematch shape:", d_rematch.shape)
```

```python
so,_ = S(structure = structure, species = species, average = "outer")
si,_ = S(structure = structure, species = species, average = "inner")
```

```python
plt.plot(np.arange(20), d_rematch[0], label='kernelPCA')
plt.plot(np.arange(20), si, label='inner')
plt.plot(np.arange(20), so, label='outer')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of Three Arrays')
plt.legend()
plt.ylim(10,-10)
plt.grid()
plt.tight_layout()
plt.show()
```
