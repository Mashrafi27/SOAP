```python
import pyzeo

structure_file = "/scratch/mmm9886/Chignolin_Trajectory/SOAP_research/CIF 166784.cif"

# First, create an AtomNetwork from your CIF file
atom_network = pyzeo.AtomNetwork.read_from_CIF(structure_file)
print("AtomNetwork created successfully")

# Define parameters
channel_radius = 1.86  # Radius to determine accessibility
probe_radius = 1.86    # Radius for MC sampling
mc_sampling_no = 5000  # Number of MC samples per atom

# 1. Calculate surface area
sa_result = pyzeo.surface_area(
    atmnet=atom_network,
    channel_radius=channel_radius,
    probe_radius=probe_radius,
    mc_sampling_no=mc_sampling_no,
    high_accuracy=False,
    exclude_pockets=True
)
print(f"Surface area result: {sa_result}")

# 2. Calculate volume
vol_result = pyzeo.volume(
    atmnet=atom_network,
    channel_radius=channel_radius,
    probe_radius=probe_radius,
    mc_sampling_no=mc_sampling_no,
    high_accuracy=False,
    exclude_pockets=True
)
print(f"Volume result: {vol_result}")

# 3. Calculate pore size distribution
# Note: This requires a histogram file path
hist_file = "psd_histogram.txt"
psd_result = pyzeo.calc_pore_size_distribution(
    atmnet=atom_network,
    channel_radius=channel_radius,
    probe_radius=probe_radius,
    mc_sampling_no=mc_sampling_no,
    hist_file=hist_file,
    high_accuracy=False,
    exclude_pockets=False
)
print(f"Pore size distribution saved to: {hist_file}")
```
